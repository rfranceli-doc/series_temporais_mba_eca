```python
#Prompt:
#acesse o https://github.com/rfranceli-doc/series_temporais_mba_eca, capture os dados do arquivo 00.BaseVolume.xlsx (Planilha2)

import pandas as pd

# Construct the raw GitHub URL for the Excel file
excel_url = 'https://raw.githubusercontent.com/rfranceli-doc/series_temporais_mba_eca/main/00.BaseVolume.xlsx'

# Read the Excel file into a pandas DataFrame, specifying the sheet name
try:
    df_volume = pd.read_excel(excel_url, sheet_name='Planilha2')
    print("Dados da Planilha2 carregados com sucesso:")
    print(df_volume.head())
except Exception as e:
    print(f"Ocorreu um erro ao carregar o arquivo ou a planilha: {e}")
```

```python
#Prompt:
#alterar os nomes dos campos, excluindo oas espaços e substituindo por _

# Replace spaces in column names with underscores
df_volume.columns = df_volume.columns.str.replace(' ', '_')

print("Nomes das colunas atualizados:")
print(df_volume.columns.tolist())
```

```python
#Prompt:
#altere o nome da coluna Usuários para Data

df_volume = df_volume.rename(columns={'Usuários': 'Data'})
print("Coluna 'Usuários' renomeada para 'Data'.")
print(df_volume.head())
```

```python
#Prompt:
#plote o campo Volume_Industrial, incorpore a quantidade de clientes industrias (Clientes_Industrial), no tempo, utilize o campo Data

import matplotlib.pyplot as plt
import pandas as pd

# Ensure 'Data' column is in datetime format
df_volume['Data'] = pd.to_datetime(df_volume['Data'])

# Create the plot
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Volume_Industrial on the first y-axis
ax1.plot(df_volume['Data'], df_volume['Volume_Industrial'], color='tab:blue', label='Volume Industrial')
ax1.set_xlabel('Data')
ax1.set_ylabel('Volume Industrial', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis for Clientes_Industrial
ax2 = ax1.twinx()
ax2.plot(df_volume['Data'], df_volume['Clientes_Industrial'], color='tab:red', label='Clientes Industrial')
ax2.set_ylabel('Clientes Industrial', color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Add title and legend
plt.title('Volume Industrial e Clientes Industrial ao longo do Tempo')
fig.tight_layout() # Adjust layout to prevent labels from overlapping

# Combine legends from both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

plt.show()
```

```python
#Prompt:
#plote o campo Volume_Cativo_-_Residencial_Coletivo, utilize no tempo o campo Data, e incorpore o campo Clientes_Cativo_-_Residencial_Coletivo

import matplotlib.pyplot as plt
import pandas as pd

# Ensure 'Data' column is in datetime format
df_volume['Data'] = pd.to_datetime(df_volume['Data'])

# Create the plot
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Volume_Cativo_-_Residencial_Coletivo on the first y-axis
ax1.plot(df_volume['Data'], df_volume['Volume_Cativo_-_Residencial_Coletivo'], color='tab:blue', label='Volume Residencial Coletivo')
ax1.set_xlabel('Data')
ax1.set_ylabel('Volume Residencial Coletivo', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis for Clientes_Cativo_-_Residencial_Coletivo
ax2 = ax1.twinx()
ax2.plot(df_volume['Data'], df_volume['Clientes_Cativo_-_Residencial_Coletivo'], color='tab:red', label='Clientes Residencial Coletivo')
ax2.set_ylabel('Clientes Residencial Coletivo', color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Add title and legend
plt.title('Volume e Clientes Residencial Coletivo ao longo do Tempo')
fig.tight_layout() # Adjust layout to prevent labels from overlapping

# Combine legends from both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

plt.show()
```

```python
#O erro que estava acontecendo
#A IA pegou a série 'volume_por_cliente_coletivo_suavizado'
#Essa séria tem NAs, por isso ela nãop calculou a suavização.
#Abaixo está a série corrigida


import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
import pandas as pd

df_clean = df_volume[['Data','volume_por_cliente_coletivo_suavizado']]
df_clean = df_clean.set_index('Data')

# Ajustar o modelo SES (Alpha = 0.2 como padrão, pode ser ajustado)
model_ses = SimpleExpSmoothing(df_clean['volume_por_cliente_coletivo_suavizado'], initialization_method='estimated').fit(smoothing_level=0.2, optimized=False)


# Adicionar os valores ajustados ao DataFrame original
df_volume['volume_por_cliente_coletivo_ses'] = df_volume['Data'].map(model_ses.fittedvalues.to_dict())

# Plotagem
plt.figure(figsize=(12, 6))
plt.plot(df_volume['Data'], df_volume['volume_por_cliente_coletivo'], label='Original (Volume por Cliente Coletivo)', color='blue', alpha=0.5)
plt.plot(df_volume['Data'], df_volume['volume_por_cliente_coletivo_ses'], label='SES (Suavizada)', color='green', linewidth=2)
plt.xlabel('Data')
plt.ylabel('Volume por Cliente')
plt.title('Série Original vs. Suavização Exponencial Simples (SES)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

```python
#Prompt:
#gere uma serie temporal suavizada utilizando ses, e plot em conjunto com a série original (volume_por cliente_coletivo)
#Usei o Gemini 3

import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
import pandas as pd

# Preparar os dados removendo NaNs para o modelo SES
df_clean = df_volume.dropna(subset=['Data', 'volume_por_cliente_coletivo']).copy()
df_clean = df_clean.set_index('Data')

# Ajustar o modelo SES (Alpha = 0.2 como padrão, pode ser ajustado)
model_ses = SimpleExpSmoothing(df_clean['volume_por_cliente_coletivo'], initialization_method='estimated').fit(smoothing_level=0.9, optimized=False)

# Adicionar os valores ajustados ao DataFrame original
df_volume['volume_por_cliente_coletivo_ses'] = df_volume['Data'].map(model_ses.fittedvalues.to_dict())

# Plotagem
plt.figure(figsize=(12, 6))
plt.plot(df_volume['Data'], df_volume['volume_por_cliente_coletivo'], label='Original (Volume por Cliente Coletivo)', color='blue', alpha=0.5)
plt.plot(df_volume['Data'], df_volume['volume_por_cliente_coletivo_ses'], label='SES (Suavizada)', color='green', linewidth=2)
plt.xlabel('Data')
plt.ylabel('Volume por Cliente')
plt.title('Série Original vs. Suavização Exponencial Simples (SES)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Verificar se há valores não-nulos agora
print(f"Valores preenchidos na série SES: {df_volume['volume_por_cliente_coletivo_ses'].notnull().sum()}")
```
